## Exercise 1 — Thresholds and error rates

Consider distinguishing $\mathcal{N}(0, 1)$ (negative) from $\mathcal{N}(1, 1)$ (positive) with the
decision rule “classify values **above** a threshold as positive.”

1. Choose a threshold that yields false positive rate $0.05$.
2. Report the corresponding true positive rate.
3. Verify your answers with `FPR_from_threshold` and `TPR_from_threshold`.


```python
dist_neg = norm(0, 1)
dist_pos = norm(1, 1)

target_fpr = 0.05
threshold = threshold_from_FPR(dist_neg, target_fpr)
tpr = TPR_from_threshold(dist_pos, threshold)

print(f"threshold = {threshold:.4f}")
print(f"TPR = {tpr:.4f}")
print(f"FPR check = {FPR_from_threshold(dist_neg, threshold):.4f}")
```
