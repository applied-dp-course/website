## Reconstruction error vs. noise (uses only public `libdpy` APIs)

```python
#| label: fig-reconstruction-error
#| fig-cap: "Reconstruction error as a function of the noise added to each query answer."
data_size = 20
noise_range = list(range(0, 11))
mechanism = QueryMechanism()

errors = []
for recon_func, name in [(lin_reg_reconstruction, "least squares"),
                         (lin_prog_reconstruction, "linear program (HiGHS)")]:
    err = run_simulation(
        data_size=data_size,
        mechanism=mechanism,
        num_experiments=5,
        recon_func=recon_func,
        query_function=choose_random_subset,
        noise_range=noise_range,
        num_queries=2 * data_size,
    )
    errors.append((err, name))

plot_reconstruction_error_as_noise_function(errors, noise_range)
```
