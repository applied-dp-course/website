## Code cells become slide output

The next cell generates a figure. Edit the Python code, rerender, and the slide image changes.

<p class="small-text">Quarto executes code cells when you render with <code>--execute</code>.</p>


````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 200)
y = 0.25 + 0.55 * x
band = 0.12

fig, ax = plt.subplots(figsize=(8, 4.2))
ax.fill_between(x, y - band, y + band, color="#dbeafe", label="accuracy slab")
ax.plot(x, y, color="#2b6cb0", linewidth=3, label="exact query answer")
ax.scatter([0, 1], [0, 1], color="#111827", s=70, zorder=3, label="Boolean corners")
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_title("A code-generated figure in a slide")
ax.grid(True, color="#e5e7eb")
ax.legend(loc="upper left")
plt.show()
````

<script>
(function () {
  const root = document.getElementById("inline-plot-demo");
  if (!root || root.dataset.ready === "true") return;
  root.dataset.ready = "true";

  const canvas = root.querySelector("#inline-plot-canvas");
  const ctx = canvas.getContext("2d");
  const noiseEl = root.querySelector("#inline-noise");
  const queriesEl = root.querySelector("#inline-queries");
  const methodEl = root.querySelector("#inline-method");
  const noiseValueEl = root.querySelector("#inline-noise-value");
  const queriesValueEl = root.querySelector("#inline-queries-value");
  const errorValueEl = root.querySelector("#inline-error-value");
  const privacyValueEl = root.querySelector("#inline-privacy-value");

  function curveValue(noise, queriesPerRecord, method) {
    const protection = noise / (noise + 8);
    const queryPressure = Math.min(1, queriesPerRecord / 8);
    const floor = method === "lp" ? 0.02 : 0.06;
    const slope = method === "lp" ? 0.62 : 0.52;
    const instability = method === "lr" ? 0.06 * Math.sin(noise / 2.6) : 0;
    return Math.max(0, Math.min(0.55, floor + slope * protection * queryPressure + instability));
  }

  function project(x, y, box) {
    return [box.left + (x / 30) * box.width, box.top + (1 - y / 0.55) * box.height];
  }

  function drawAxes(box) {
    ctx.strokeStyle = "#1f2933";
    ctx.lineWidth = 2;
    ctx.beginPath();
    ctx.moveTo(box.left, box.top);
    ctx.lineTo(box.left, box.top + box.height);
    ctx.lineTo(box.left + box.width, box.top + box.height);
    ctx.stroke();

    ctx.fillStyle = "#5b6773";
    ctx.font = "13px system-ui";
    ctx.fillText("noise scale", box.left + box.width - 80, box.top + box.height + 42);
    ctx.save();
    ctx.translate(box.left - 48, box.top + 112);
    ctx.rotate(-Math.PI / 2);
    ctx.fillText("average reconstruction error", 0, 0);
    ctx.restore();

    ctx.strokeStyle = "#e5e7eb";
    ctx.lineWidth = 1;
    for (let i = 0; i <= 5; i++) {
      const y = i * 0.1;
      const [, py] = project(0, y, box);
      ctx.beginPath();
      ctx.moveTo(box.left, py);
      ctx.lineTo(box.left + box.width, py);
      ctx.stroke();
      ctx.fillText(y.toFixed(1), box.left - 34, py + 4);
    }
    for (let x = 0; x <= 30; x += 5) {
      const [px] = project(x, 0, box);
      ctx.beginPath();
      ctx.moveTo(px, box.top);
      ctx.lineTo(px, box.top + box.height);
      ctx.stroke();
      ctx.fillText(String(x), px - 6, box.top + box.height + 20);
    }
  }

  function drawCurve(box, queriesPerRecord, method, color, width) {
    ctx.strokeStyle = color;
    ctx.lineWidth = width;
    ctx.beginPath();
    for (let noise = 0; noise <= 30; noise += 0.25) {
      const [x, y] = project(noise, curveValue(noise, queriesPerRecord, method), box);
      if (noise === 0) ctx.moveTo(x, y);
      else ctx.lineTo(x, y);
    }
    ctx.stroke();
  }

  function draw() {
    const selectedNoise = Number(noiseEl.value);
    const queriesPerRecord = Number(queriesEl.value);
    const method = methodEl.value;
    const box = { left: 78, top: 34, width: canvas.width - 128, height: canvas.height - 118 };
    const highlighted = curveValue(selectedNoise, queriesPerRecord, method);
    const privacyProxy = selectedNoise / (selectedNoise + 11);

    ctx.clearRect(0, 0, canvas.width, canvas.height);
    ctx.fillStyle = "#fff";
    ctx.fillRect(0, 0, canvas.width, canvas.height);
    drawAxes(box);

    drawCurve(box, queriesPerRecord, "lr", "#d62728", method === "lr" ? 4 : 2);
    drawCurve(box, queriesPerRecord, "lp", "#1f77b4", method === "lp" ? 4 : 2);

    const [px, py] = project(selectedNoise, highlighted, box);
    ctx.strokeStyle = "#6b7280";
    ctx.lineWidth = 1.5;
    ctx.setLineDash([6, 6]);
    ctx.beginPath();
    ctx.moveTo(px, box.top);
    ctx.lineTo(px, box.top + box.height);
    ctx.stroke();
    ctx.setLineDash([]);

    ctx.beginPath();
    ctx.arc(px, py, 8, 0, Math.PI * 2);
    ctx.fillStyle = method === "lp" ? "#1f77b4" : "#d62728";
    ctx.fill();
    ctx.strokeStyle = "#111827";
    ctx.lineWidth = 2;
    ctx.stroke();

    ctx.fillStyle = "#1f2933";
    ctx.font = "16px system-ui";
    ctx.fillText("LP", box.left + box.width - 80, project(30, curveValue(30, queriesPerRecord, "lp"), box)[1] - 8);
    ctx.fillText("LR", box.left + box.width - 80, project(30, curveValue(30, queriesPerRecord, "lr"), box)[1] + 22);

    noiseValueEl.textContent = String(selectedNoise);
    queriesValueEl.textContent = String(queriesPerRecord);
    errorValueEl.textContent = highlighted.toFixed(2);
    privacyValueEl.textContent = privacyProxy.toFixed(2);
  }

  noiseEl.addEventListener("input", draw);
  queriesEl.addEventListener("input", draw);
  methodEl.addEventListener("change", draw);
  draw();
})();
</script>

## Code cell options

Add Quarto options as comments at the top of a code cell:

```python
#| echo: false
#| fig-width: 8
#| fig-height: 4

plot_my_figure()
```

- `echo: false` hides source code and shows output only.
- `eval: false` shows code without running it.
- `fig-width` and `fig-height` control generated figure size.
